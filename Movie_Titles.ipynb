```python
# Read csv file 
csv_file = 'Resources/title_basics.csv'
title_basics = pd.read_csv(csv_file)
```

```python
# Create df from csv
title_basics_df = pd.DataFrame(title_basics)
title_basics_df.head()
```

```python
# Drop the rows that have N/A for startYear
clean_title_basics_df = title_basics_df[pd.to_numeric(title_basics_df['startYear'], errors='coerce').notnull()]
# Filter df to just releases >= 2000
recentTitles_df = clean_title_basics_df.loc[clean_title_basics_df['startYear'].astype('int32') >= 2000,:]
recentTitles_df.head()
```

```python
# Types of Titles in df
recentTitles_df['titleType'].value_counts()
```

```python
# Filter df to just Movies & remove Adult Movies
recentMovies_df = recentTitles_df.loc[(recentTitles_df['titleType'] == 'movie') & (recentTitles_df['isAdult'] == 0),:]
recentMovies_df.head()
```

```python
# See difference between primaryTitle and originalTitle
recentTitles_df.loc[recentTitles_df['primaryTitle'] != recentTitles_df['originalTitle'],:]
```

```python
# List of just Movie Titles using primaryTitle
recentMovieTitles = [i for i in recentMovies_df['primaryTitle']]
# Total Movie Titles in List
print(f'''-----------------------
Total # Movies: {len(recentMovieTitles)}
-----------------------''')
# Preview List
print('''
------
MOVIES
------''')
for movie in recentMovieTitles[:15]:
    print(movie)
```

```python
# Save Cleaned Movie List to csv
recentMovieTitles_df = pd.DataFrame(recentMovies_df.loc[:,['primaryTitle','startYear','genres']])
recentMovieTitles_df.to_csv('Resources/recentMovieTitles.csv', index=False)
recentMovieTitles_df.to_csv('Resources/recentMovieTitles.csv', mode='a', header=False index=False)
```
